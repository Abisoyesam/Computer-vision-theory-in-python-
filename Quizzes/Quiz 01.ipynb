## Q2: Create an RGB and BGR image

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rgb = np.zeros((100,150,3), dtype='uint8')
plt.imshow(rgb)
```

```python
rgb[:,0:50,0] = 255 # Red channle
rgb[:,50:100,1] = 255 # Green channel
rgb[:,100:,2] = 255 # Blue channel

plt.imshow(rgb);
```

```python
# Plot BGR image

plt.imshow(rgb[:,:,::-1]);
```

## Q3: Extract yellow roses from the following image.

<img src="../images/tulips.jpg" alt="drawing" width="320"/>

**Hint:** Hsv range Lower range is (20,100,100) and upper range is (30,255,255)

```python
import cv2
```

```python
img = cv2.imread('../images/tulips.jpg')
imgHsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Yellow rose

ly = np.array([20,100,100])
uy = np.array([30,255,255])

mask = cv2.inRange(imgHsv, ly,uy)
```

```python
res = cv2.bitwise_and(img,img,mask=mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
