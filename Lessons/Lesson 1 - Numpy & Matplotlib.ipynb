## Matplotlib

It is a python plotting library that allows to visualize data. Matplotlib has many visualization features similiar to MatLab. 

* Built on Numpy arrays (and Python)
* Integrates directly with pandas
* Can create basic or advanced plots
* It is for plot 2D graph with limited support of 3D graph

### Workflow in Matplotlib
1. Create data
1. Create plot (figure)
1. Plot data (axes on figure)
1. Customise plot (adding label, title and so on)
1. Save or share plot

* Matplotlib has two interfaces that are used for ploting:
    1. Stateful Interface: it uses `pyplot` class and keep track of current state in the object.
    1. Object oriented interface: it uses a hierachy of objects to represent various object of plot. Two main objects used are `figure` and `axes`.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Stateful Interface
This interface uses the pyplot class, which is based on Matlab. A single object of this class is created, which is used for all plotting purposes.

```python
plt.plot();
```

```python
plt.plot()

# Notice that without the semi-colon, we have empty list.
```

```python
x = [1,2,3,4]
y = [3,6,9,12]
```

```python
plt.plot(x,y);
```

## Object oriented Interface

In this interface, we use different objects for different elements of the plot. The two main objects that are used in this interface for plotting are:
* The **figure object**, which acts as the container for other objects.
* The **axes object**, which is the actual plotting area containing the x axis, y axis, points, lines, legends, and labels. Note that the axes here does not refer to the x and y axes but the entire subplot.

```python
# subplots method create a tuple returning a figure and one or more axes 
fig,ax = plt.subplots()
ax.plot(x,y);
```
