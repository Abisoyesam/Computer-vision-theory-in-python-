## Step 1: Gaussian Filter

$
{\Large G(x,y) = \dfrac{1}{2\pi \sigma^2}e^{-\dfrac{x^2 + y^2}{2\sigma^2}}}
$

Assuming $\sigma = 1$ 
![gauss](../images/gausmask.png)

```python
def G(x,y):
    sigma = 1
    den = 2 * np.pi * sigma**2
    oth = -(x**2 + y**2)/(2*sigma**2)
    exp = np.exp(oth)
    return 1/den*exp
```

**Computing all the Gaussian will help to determine the mask but how do we determine the value of $\sigma$ and the size of filter ?**

Should it be $3x3$ or $5x5$ and so on.

![Guassian filter](../images/gausfilter.png)

The sigma actually controls the width of the filter(kernel or mask). If sigma increase, the peak decreases. 

```python
def f_getGuassianFilter(s,sigma):
    # s is the size of the window
    maxX = s//2 
    minX = -maxX 
    minY = minX 
    maxY = maxX
    G_arr = np.zeros((s,s))
    for x in range(minX,maxX+1):
        for y in range(minY,maxY+1):
            v = G(x,y)
            G_arr[x-minX,y-minY] = v
    return G_arr
```

```python
G = f_getGuassianFilter(5,1)
```

```python
G
```

```python
np.sum(G)
```

**N.B:** If $\sigma = 1$, the best window size is **5** and if $\sigma = 2$ then window size should be atleast **11** ie (11 x 11) filter

### Derivative of Gaussian (D.O.G)

Computing the **derivative of the smoothing filters** can be much effecient to compute the final derivative as compared to applying the smoothing filter first and then computing the derivative.

$
{\Large G(x,y) = \dfrac{1}{2\pi \sigma^2}e^{-\dfrac{x^2 + y^2}{2\sigma^2}}}
$

$G_x = \dfrac{\delta G_{\sigma}(x,y)}{\delta x} = \dfrac{-x}{2\pi\sigma^4}exp\left(-\dfrac{x^2 + y^2}{2\sigma^2}\right) \\[7mm] = \dfrac{-x}{\sigma^2}\left(\dfrac{1}{2\pi \sigma^2}e^{-\dfrac{x^2 + y^2}{2\sigma^2}}\right)$

$G_y = \dfrac{\delta G_{\sigma}(x,y)}{\delta x} = \dfrac{-y}{2\pi\sigma^4}exp\left(-\dfrac{x^2 + y^2}{2\sigma^2}\right) \\[7mm] = \dfrac{-y}{\sigma^2}\left(\dfrac{1}{2\pi \sigma^2}e^{-\dfrac{x^2 + y^2}{2\sigma^2}}\right)$

```python
def Gx(x,y):
    sigma = 1
    den = 2 * np.pi * sigma**4
    oth = -(x**2 + y**2)/(2*sigma**2)
    exp = np.exp(oth)
    return -x/den*exp
```

```python
def GaussianDx(s,sigma):
    # s is the size of the window
    maxX = s//2 
    minX = -maxX 
    minY = minX 
    maxY = maxX
    G_x = np.zeros((s,s))
    for x in range(minX,maxX+1):
        for y in range(minY,maxY+1):
            v = Gx(x,y)
            G_x[x-minX,y-minY] = v
    return G_x
```

```python
def Gy(x,y):
    sigma = 1
    den = 2 * np.pi * sigma**4
    oth = -(x**2 + y**2)/(2*sigma**2)
    exp = np.exp(oth)
    return -y/den*exp
```

```python
def GaussianDy(s,sigma):
    # s is the size of the window
    maxX = s//2 
    minX = -maxX 
    minY = minX 
    maxY = maxX
    G_y = np.zeros((s,s))
    for x in range(minX,maxX+1):
        for y in range(minY,maxY+1):
            v = Gy(x,y)
            G_y[x-minX,y-minY] = v
    return G_y
```

## Step 4: Non-Maxima Suppression (NMS)

![maxima](../images/maxima.png)

```python
plt.hist(Gd.flatten())
```

**Gradient Direction Quantization Implementation**

```python
bins = np.array([-180.,-135.,-90.,-45.,0.,45.,90.,135.,180.])
# -1 is to make it zero indexing
indx = np.digitize(Gd, bins)-1
Gd_bin = bins[indx.flatten()].reshape(Gd.shape)
```

```python
i,j = 50,150
Gd[i,j], Gd_bin[i,j]
```

```python
# use of digitize
A = np.array([134,179,-33])
indx = np.digitize(A,bins)
```

```python
# return nearest value (- or +) in the bin
indx
```

**NMS Implementation**

```python
def NMS(Gm,Gd):
    Gm_NMS = np.zeros(Gm.shape)
    numRows,numCols = Gm.shape[0], Gm.shape[1]
    a,b = 0.,0.
    for r in range(1,numRows-1):
        for c in range(1,numCols-1):
            ang = Gd_bin[r,c]
            if ang == 180. or ang == -180. or ang == 0.0:
                a,b = Gm[r-1,c], Gm[r+1,c]
            elif ang == 45. or ang == -135.:
                a,b = Gm[r+1,c+1],Gm[r-1,c-1]
            elif ang == 90. or ang == -90.:
                a,b = Gm[r,c-1],Gm[r,c+1]
            elif ang == 135. or ang == -45.:
                a,b = Gm[r-1,c+1], Gm[r+1,c-1]
            else:
                print("Error")
                return
            if Gm[r,c] >= a and Gm[r,c] >= b:
                Gm_NMS[r,c] = Gm[r,c]
    return Gm_NMS
```

```python
Gm_NMS = NMS(Gm,Gd)
```

```python
plt.subplot(121)
plt.imshow(Gm, cmap='gray')
plt.title('Gm')
plt.axis('off')
plt.subplot(122)
plt.title('Gm_NMS')
plt.imshow(Gm_NMS, cmap='gray');
plt.axis('off');
```

```python
plt.imshow(Gm_NMS>25, cmap='gray')
```

## Step 5: Hesterysis Thresholding

```python
from skimage import filters
```

```python
# L is low threshold
L = Gm_NMS.mean()

# H is High threshold
H = Gm_NMS.std() + L

# E is edge image
E = filters.apply_hysteresis_threshold(Gm_NMS,L,H)
```
