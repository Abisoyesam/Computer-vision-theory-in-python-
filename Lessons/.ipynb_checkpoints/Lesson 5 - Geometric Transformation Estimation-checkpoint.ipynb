### Affine transformation VS Projective transformation

**Affine transformation:** It is otherwise known as homogenous transformation while **Projective transformation** is known as homography transformation.

The difference between both can be explained using the transformation matrix. **The third row of the transformation matrix is the distinguishable difference.**

Affine: $\begin{bmatrix} a&b&c \\ d&e&f \\ 0&0&1\end{bmatrix}$

Projective: Affine: $\begin{bmatrix} a&b&c \\ d&e&f \\ 2&0.2&0.1\end{bmatrix}$

**N.B:** In affine the third row of the matrix is nothing but $\begin{bmatrix}0&0&1\end{bmatrix}$, but for projective, the matrix is can be $\begin{bmatrix}0&0&1\end{bmatrix}$ or anything.

Hence, **Affine $\subset$ Projective**

```python
# display image as original

import matplotlib as mpl

def displayImageInActualSize(I):
    dpi = mpl.rcParams['figure.dpi']
    H,W = I.shape
    figSize = W/float(dpi), H/float(dpi)
    fig = plt.figure(figsize=figSize)
    ax = fig.add_axes([0,0,1,1])
    ax.axis('off')
    ax.imshow(I,cmap='gray')
    plt.show
```

```python
# bilinear interpolation

def bilinearInterpolate(row,col,I):
    lc = int(col)
    rc = lc + 1
    wr = col - lc
    wl = rc - col
    tr = int(row)
    br = tr + 1
    wt = br - row
    wb = row - tr
    if tr >= 0 and br < I.shape[0] and lc >= 0 and rc < I.shape[1]:
        a = wl * I[tr,lc] + I[tr,rc]
        b = wl * I[br,lc] + wr*I[br,rc]
        g = wt * a + wb * b

        return np.uint8(g)
    else:
        return 0
```

```python
def f_getExtentsProjective(T,rMax,cMax):
    # all the cordinates are now homogeneous by adding 1
    cords = np.array([[0,0,1],[0,cMax-1,1],[rMax-1,0,1],[rMax-1,cMax-1,1]])
    A_dash = T.dot(cords.T)
    # the modification
    A_dash = A_dash/A_dash[2,:]
    mins = A_dash.min(axis=1)
    maxs = A_dash.max(axis=1)
    minR = np.int64(np.floor(mins[0]))
    minC = np.int64(np.floor(mins[1]))
    maxR = np.int64(np.ceil(maxs[0]))
    maxC = np.int64(np.ceil(maxs[1]))
    H,W = maxR - minR + 1, maxC - minC + 1
    return minR, minC, maxR, maxC, H, W
```

```python
def f_transformProjective(T, img):
    rMax,cMax = img.shape[0],img.shape[1]
    minR, minC, maxR, maxC, H, W = f_getExtentsProjective(T,rMax,cMax)
    I2 = np.zeros((H,W),dtype="uint8")
    Tinv = np.linalg.inv(T)
    for new_i in range(minR,maxR):
        for new_j in range(minC,maxC):
            # convert P_dash to homogeneous
            P_dash = np.array([new_i,new_j,1])
            P = Tinv.dot(P_dash)
            # modification
            P = P/P[2]
            i,j = P[0], P[1]
            if i < 0 or i >= rMax or j < 0 or j >= cMax:
                pass
            else:
                g = bilinearInterpolate(i,j,img)
                I2[new_i-minR,new_j-minC] = g
                
    return I2
```

## 2. Projective (Homography) Transformation Estimation

$\begin{bmatrix}x'\\y'\\1\end{bmatrix} = {\LARGE \alpha} \begin{bmatrix}h1&h2&h3\\h4&h5&h6\\h7&h8&h9\end{bmatrix} \begin{bmatrix}x \\y\\1\end{bmatrix}$

**Direct Linear Transform (DLT): the algorithm use to estimate homography**

$\begin{equation}x' = \alpha (h_1x + h_2y + h_3)\\y' = \alpha (h_4x + h_5y + h_6)\\ 1 = \alpha(h_7x + h_8y + h_9)\end{equation}$

$\alpha = \dfrac{1}{h_7x + h_8y + h_9}\\[10mm] x' = \dfrac{h_1x + h_2y + h_3}{h_7x + h_8y + h_9}\\[10mm] y' = \dfrac{h_4x + h_5y + h_6}{h_7x + h_8y + h_9}\\[10mm]$

$x'(h_7x + h_8y + h_9) = (h_1x + h_2y + h_3)\\[4mm] y'(h_7x + h_8y + h_9) = (h_4x + h_5y + h_6)\\[10mm]$

$h_7xx' + h_8yx' + h_9x' - h_1x - h_2y - h_3 = 0\\[4mm] h_7xy' + h_8yy' + h_9y' - h_4x -h_5y - h_6 = 0$

![DLT](../images/DLT.png)

**h:** right null vector.

#### Implementing DLT

```python
# compute the nul value

def f_computeRightNullSpace(A):
    B = A.T.dot(A)
    w,v = np.linalg.eig(B)
    idx = np.argmin(w)
    nullVector = v[:,idx]
    return nullVector
```

## 3. Panorama Stitching

```python
img1 = cv2.imread('../images/im1.PNG')
img2 = cv2.imread('../images/im2.PNG')
img3 = cv2.imread('../images/im3.PNG')
```

```python
imgs = []
imgs.append(img1)
imgs.append(img2)
imgs.append(img3)
```

**N.B: The list imgs does not necessarily has to be in any order.**

```python
# stitcher module

M = cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
status, panorama = M.stitch(imgs)
```

```python
%matplotlib inline
plt.imshow(panorama[:,:,::-1]);
```
