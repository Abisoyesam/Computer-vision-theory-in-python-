### Convolution vs Image Filtering

Convolution is the same as image filtering. Somehow, Convolution gains more popularity than image filtering but mostly in computer vision and convolution neural network (CNN), the operation convolution is the same as filtering.

* Cross corelation: filter placed on image then dot product of coinciding patch
* Image filtering: applying cross corelation on every pixel of image.
* Convolution: before applying cross corelation flip the mask both horizontal and vertically.

## 3. Image Sharpening

![image](../images/sharp.png)

It is the reverse process of image blurring.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
im = plt.imread('../images/bird (1).jpeg')
```

```python
im.shape
```

```python
imGray = cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
```

```python
smootMask = np.ones((15,15))/255
```

**Image Blurring**

```python
# boundary: symmetrical
# mode: same

imBlurred = signal.convolve2d(imGray,smootMask,boundary='symm', mode='same')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,15))
ax1.imshow(imGray, cmap="gray")
ax2.imshow(imBlurred, cmap='gray')
```

**Edge detection**

```python
xMask = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
yMask = xMask.T.copy()
```

```python
fx = signal.convolve2d(imGray,xMask,boundary='symm', mode='same')
fy = signal.convolve2d(imGray,yMask,boundary='symm', mode='same')
```

```python
# Gradient magnitude

Gm = (fx**2 + fy**2)**0.5
```

```python
plt.imshow(Gm, cmap='gray')
```

**Image Shapening**

```python
imSharp = imBlurred + 0.01*Gm
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,15))
ax1.imshow(imBlurred,cmap='gray')
ax2.imshow(imSharp,cmap='gray')
```

## 4. Low-Pass and High-Pass

Low pass filter allows only low frequency from the filter. It smooths out.

High pass filter sharpen edge detection.

Band pass filter: do not pick very low or very high frequency but just the medium.
