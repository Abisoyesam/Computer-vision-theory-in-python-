## Convolutional Neural Networks

Object detection does not really depend on classical method (hand engineering) but rather deep neural network (CNN).

**YOLO:** is based on convolution Neural Networks. For a layman, CNNs can be seen as feature extractor & classifier. (You Only Look Once)

YOLO is not only for person detection but all common type of objects.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

## RCNN

**RCNN:** Region proposal Convolution Neural Network. It can be used instead of Yolo for object detection and sometimes gives a better result.

1. Image is first segmented
2. Then classification

RCNN = Segmentation + Classification
