# Lesson 6: Binary Morphology

**Binary Image:** A binary image is like a grayscale image but each pixel value is either zero (0) or one (1).

Zero - Non existence, One -  Existence

**Morphological transformation:** transformation like erosion, dilation etc

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../images/ein.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray');
```

```python
# binary image
imgB = cv2.imread("../images/binary.PNG",cv2.IMREAD_GRAYSCALE)
plt.imshow(imgB, cmap='gray');
```

```python
# np.unique(img)

# if the image is binary, the uniqueness will either be 0 or 1 
# but if grayscale image 0 - 255
```

```python
# Structuring element
# kernel retangle

k_rect = cv2.getStructuringElement(cv2.MORPH_RECT,(15,15))
```

```python
type(k_rect)
```

```python
k_rect.dtype
```

```python
k_rect
```

```python
k_elipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(15,15))
k_elipse
```

```python
k_cross = cv2.getStructuringElement(cv2.MORPH_CROSS,(15,15))
k_cross
```

## 1. Erosion Theory

![erosion](../images/erosion.png)

Erosion removes some of the pixels at certain bondaries that are unwanted. You take the 'AND' of structuring elements.

```python
# Erosion with kernel

E_r = cv2.erode(imgB,k_rect)
plt.imshow(E_r, cmap='gray')
```

```python
E_e = cv2.erode(imgB,k_elipse)
plt.imshow(E_e, cmap='gray')
```

```python
E_c = cv2.erode(imgB,k_cross)
plt.imshow(E_c, cmap='gray')
```

```python
# although visually they look the same but they aint dsame
# using eucludean distance

print(np.sum((E_r - E_e)**2))
```

```python
print(np.sum((E_r - E_c)**2))
```

```python
print(np.sum((E_e - E_c)**2))
```

## 2. Dilation

![dilation](../images/dilation.png)

Dilation: it adds pixel element. i.e Taking the `OR` of the structure element.
**N.B:** OR (+) and AND (x)

```python
D_r = cv2.dilate(imgB,k_rect)
plt.axis('off')
plt.imshow(D_r, cmap='gray')
```

```python
D_e = cv2.dilate(imgB,k_elipse)
plt.axis('off')
plt.imshow(D_e, cmap='gray')
```

```python
D_c = cv2.dilate(imgB,k_cross)
plt.axis('off')
plt.imshow(D_c, cmap='gray')
```

```python
print(np.sum((D_r - D_e)**2))
```

## 3. Image Opening

![open](../images/open.png)

It can be used to remove noise from the image and still maintain the image

```python
imgB.dtype
```

```python
im_bool = np.array(imgB[:,:], dtype=bool)
```

```python
im_bool.dtype
```

```python
plt.imshow(im_bool,cmap='gray')
```

```python
noise = np.random.rand(im_bool.shape[0],im_bool.shape[1])
```

```python
N_bin = noise > 0.95
```

```python
plt.imshow(N_bin, cmap='gray')
```

```python
noisyImg = im_bool | N_bin
```

```python
plt.imshow(noisyImg, cmap='gray')
```

**Remove Noise**

```python
# original image + rect kernel

O_r = cv2.morphologyEx(np.float32(noisyImg), cv2.MORPH_OPEN, k_rect)
```

```python
# You can see that the image noise has been removed.

plt.imshow(O_r, cmap='gray')
```

**Remove noise by Erosion followed by Dilation**

```python
E_r = cv2.erode(np.float32(noisyImg),k_rect)
O_r2 = cv2.dilate(E_r,k_rect)
plt.imshow(O_r2, cmap='gray')
```

## 4. Image Closing

![close](../images/close.png)

It handles holes filling.

```python
N_bin = noise < 0.95
plt.imshow(N_bin,cmap='gray')
```

```python
holesImg = im_bool & N_bin
plt.axis('off')
plt.imshow(holesImg,cmap='gray')
```

**Fill holes**

```python
# original image + rect kernel

C_r = cv2.morphologyEx(np.float32(holesImg), cv2.MORPH_CLOSE, k_rect)
```

```python
plt.imshow(C_r, cmap='gray')
```

**Filling holes by dilating followed by erosion**

```python
D_r = cv2.dilate(np.float32(holesImg),k_rect)
O_r2 = cv2.erode(D_r,k_rect)
plt.imshow(O_r2, cmap='gray')
```

## 5. Gradient

![gradient](../images/grad.png)

First dilate (expand) then substract erosion (shrink).

```python
G_r = cv2.morphologyEx(imgB, cv2.MORPH_GRADIENT, k_rect)
plt.imshow(G_r, cmap='gray')
```
