## Color Space (HSV)

* **H** is the Hue/color in degree $\in [0,360]$
* **S** is the Saturation $\in [0,1]$
* **V** is the Value, Brightness or Luminance $\in [0,1]$

**Conversion RGB $\to$ HSV**
* $\dfrac{R, G, B}{255}$


* V is the max of R,G,B i.e `max(R,G,B)`, but you will also compute the minimum value i.e `min(R,G,B)`


* S = $\dfrac{(max - min)}{max}$  (or S = 0, if V = 0)


* H = 60 $\times \left\{ \begin{array}{11}
        0 + (G - B) / (max-min), & \mbox{if max = R} \\ 
        2 + (B - R) / (max-min), & \mbox{if max = G} \\
        4 + (R - G) / (max-min), & \mbox{if max = B} \end{array} \right.$
        
  H = H + 360, if H < 0
  
**N.B:** HSV is more close to how human vision produces or perceives different color attributes.

### RGB $\to$ HSV using python

```python
def RGB2HSV(R,G,B, scaleFactor):
    '''
    Conversion of RGB color space to HSV
    ScaleFactor: To change the range of the Value and Saturation
    '''
    r,g,b = R/255, G/255, B/255
    vmax = max(r,g,b) # maximum of r,g,b
    vmin = min(r,g,b) # minimum of r,g,b
    diff = vmax - vmin # difference of vmax and vmin
    
    if vmax == vmin:
        h = 0
    elif vmax == r:
        h = (60 * ((g-b)/diff) + 0) % 360
    elif vmax == g:
        h = (60 * ((b-r)/diff) + 120) % 360
    elif vmax == b:
        h = (60 * ((r-g)/diff) + 240) % 360
        
    if h < 0:
        h = h + 360
    if vmax == 0:
        s = 0
    else:
        s = (diff/vmax) * scaleFactor
    
    v = scaleFactor * vmax
    return h, s, v
```

```python
print(RGB2HSV(100,200,50,100))
```

## Extraction or Segmentation using HSV

In Hsv color shape segmentation come in handy and easy to do.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/tulips.jpg')
imgHsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# RGB image
plt.imshow(img[:,:,::-1]);
```

```python
# Hsv image
plt.imshow(imgHsv);
```

**N.B:** 
1. Hue values are distributed over a circle (range btw 0 - 360 degree) but in Opencv to fit into 8bits value the range is from 0 - 180.
1. The red color, in OpenCV, has the hue values approximately in the range of **0 to 10** and **160 to 180**.
1. We use this range to avoid detection of skin as red.

```python
# segment the red roses
# [h, s, v]

lr = np.array([0,120,70]) #lower red
ur = np.array([10,255,255]) #Upper red

mask1 = cv2.inRange(imgHsv,lr,ur)
```

```python
lr = np.array([160,120,70]) #lower red
ur = np.array([180,255,255]) #Upper red

mask2 = cv2.inRange(imgHsv,lr,ur)
```

```python
mask = mask1 | mask2
```

```python
np.unique(mask)
```

```python
# resultant image

res = cv2.bitwise_and(img,img,mask=mask)
```

```python
# detect the red

plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
